# 정사각형의 개수
한 변의 길이가 N인 정사각형 모양의 격자판이 있다.<br>
이 격자판의 각 격자는 길이가 1일 때, 이 격자판의 크고 작은 정사각형의 개수는 총 몇 개일까?
<br><br>
예를 들어 위 그림에서 보이는 한 변의 길이가 4인 정사각형 격자판에서 찾을 수 있는 크고 작은 정사각형은 총 30개이다. <br>
이와 같이 정사각형 격자판의 한 변의 길이 N이 입력을 통해 주어질 때, 이 격자판에서 찾을 수 있는 모든 정사각형의 개수를 구하여라.<br><br>

## 입력
정사각형 격자판의 한 변의 길이를 나타내는 정수 N이 주어진다.
    <br>
## 출력
격자판에서 찾을 수 있는 모든 정사각형의 개수를 출력한다.
  > 출처: 구름LEVEL 알고리즘 프로그래밍 문제

```python
# -*- coding: utf-8 -*-
# UTF-8 encoding when using korean
n = int(input())
a = 0
for i in range(n+1):
	a = a + i**2
print(a)
```
